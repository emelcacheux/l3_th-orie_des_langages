Etape 1 : Définissez les différents token en leur donnant un numéro

```python
/*Les e.r.*/

lettres [a b c]
epsilon [E]
operateur [* . +]

/*Tokens pour les parenthèses*/

#define PAR_O 101 // '(' parenthèse ouvrante 
#define PAR_F 102 // ')' parenthèse fermante

/*Tokens pour les opérateurs*/

#define PLUS 201 // '+'
#define ETOILE 202 //'*'
#define POINT 203 // '.'

/*Tokens pour les symboles*/

#define VARIABLE 301 // '[a b c]'
#define EPSILON 302 // 'E'
```

Etape 2 : Créez un fichier lex regexp.l qui permettra de reconnaître les différents tokens


// Fichier token.h

#define PAR_O 101
#define PAR_F 102

#define PLUS    201
#define ETOILE  202
#define POINT   203

#define VARIABLE 301
#define EPSILON  302

```python
// Fichier regexp.l

%{
#include <stdio.h>
#include "token.h"
%}

/* EXPRESSIONS RÃ‰GULIÃˆRES FLEX */
lettres      [abc]
epsilon      E
operateur    [*.+]

%%

"("         { return PAR_O; }
")"         { return PAR_F; }
"+"         { return PLUS; }
"*"         { return ETOILE; }
"."         { return POINT; }

{epsilon}   { return EPSILON; }
{lettres}   { return VARIABLE; }

[ \t\r\n]+  ;

%%
int main() {
    int tok;
    while ((tok = yylex()) != 0) {
        printf("Token: %d\n", tok);
    }
    return 0;
}
```

Etape 5 : Créez un fichier yacc regexp.y contenant les règles de votre grammaire

```python
%% 
%token <str> VARIABLE
%token EPSILON
       ETOILE
       PLUS
       PAR_O
       PAR_F
%LEFT PLUS 
%LEFT POINT
%RIGHT ETOILE
%%

%%
Expression : VARIABLE
    | EPSILON
    | expression ETOILE
    | expression PLUS Expression
    | expression POINT expression
    | PAR_O expression PAR_F
; 
%%
```
